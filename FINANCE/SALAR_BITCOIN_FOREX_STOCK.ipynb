```python
def ROC(data):# calculate rate of change
    roc = [0]
    for i in range(1, len(data)):
        roc.append((data[i] - data[i-1]) / data[i-1]) 
    return 100*roc
```

```python
### PLOT PRICE CANDLESTICKS
# Define a function to plot a candlestick chart
def plot_candlestick(ticker, start, end):
    import matplotlib.pyplot as plt
    import yfinance as yf
    # Get the data from yfinance
    data = yf.download(ticker, start, end)
    # Create a figure and an axis
    fig, ax = plt.subplots(figsize=(12,8))
    # Set the title and the labels
    ax.set_title(f"{ticker} Candlestick Chart from {start} to {end}")
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    # Loop through the data and plot the candlesticks
    for i in range(len(data)):
        # Get the open, high, low and close prices for each day
        Open = data["Open"][i]
        high = data["High"][i]
        low = data["Low"][i]
        close = data["Close"][i]
        # Get the date as a string
        date = data.index[i].strftime("%Y-%m-%d")
        # Determine the color and the width of the candlestick
        if close > Open:
            color = "green"
            width = 0.5
        elif close < Open:
            color = "red"
            width = -0.5
        else:
            color = "black"
            width = 0.1
        # Plot the candlestick body
        ax.bar(date, close - Open, bottom=Open, color=color, width=width)
        # Plot the candlestick shadow
        ax.vlines(date, low, high, color=color)
    # Show the plot
    plt.xticks(rotation=90)
    plt.semilogy()
    #plt.grid()
    plt.show()

plot_candlestick(NAME, START, END)  
```

```python
def plot_heikin_ashi(ticker, start, end):
    import matplotlib.pyplot as plt
    import yfinance as yf
    import pandas as pd
    # Get the data from yfinance
    data = yf.download(ticker, start, end)
    # Calculate the Heikin-Ashi values
    ha_data = pd.DataFrame(index=data.index, columns=["Open", "High", "Low", "Close"])
    ha_data["Close"] = (data["Open"] + data["High"] + data["Low"] + data["Close"]) / 4
    ha_data["Open"] = 0.5 * (data["Open"].shift(1) + data["Close"].shift(1))
    ha_data["Open"][0] = (data["Open"][0] + data["Close"][0]) / 2
    ha_data["High"] = data[["Open", "High", "Low", "Close"]].max(axis=1)
    ha_data["Low"] = data[["Open", "High", "Low", "Close"]].min(axis=1)
    # Create a figure and an axis
    fig, ax = plt.subplots(figsize=(12,8))
    # Set the title and the labels
    ax.set_title(f"{ticker} Heikin-Ashi Candlestick Chart from {start} to {end}")
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    # Loop through the data and plot the candlesticks
    for i in range(len(ha_data)):
        # Get the open, high, low and close prices for each day
        Open = ha_data["Open"][i]
        high = ha_data["High"][i]
        low = ha_data["Low"][i]
        close = ha_data["Close"][i]
        # Get the date as a string
        date = ha_data.index[i].strftime("%Y-%m-%d")
        # Determine the color and the width of the candlestick
        if close > Open:
            color = "green"
            width = 0.5
        elif close < Open:
            color = "red"
            width = -0.5
        else:
            color = "black"
            width = 0.1
        # Plot the candlestick body
        ax.bar(date, close - Open, bottom=Open, color=color, width=width)
        # Plot the candlestick shadow
        ax.vlines(date, low, high, color=color)
    # Show the plot
    plt.xticks(rotation=90)
    plt.semilogy()
    #plt.grid()
    plt.show()

plot_heikin_ashi(NAME, START, END)   
```

```python
# Define a function to calculate the volume by price indicator
def volume_by_price(ticker, start, end, bins):
    # Import yfinance and matplotlib libraries
    import yfinance as yf
    import matplotlib.pyplot as plt  
    # Get the historical data for the ticker
    data = yf.download(ticker, start, end)
    # Calculate the volume-weighted average price (VWAP)
    data['VWAP'] = (data['Adj Close'] * data['Volume']).cumsum() / data['Volume'].cumsum()
    # Group the data by VWAP bins and sum the volume
    groups = data.groupby(pd.cut(data['VWAP'], bins))
    vbp = groups['Volume'].sum()
    # Return the vbp series
    return vbp

# Example: plot the volume by price indicator for AAPL from 2020-01-01 to 2020-12-31 with 10 bins
vbp = volume_by_price(NAME, START, END, 30)
plt.figure(figsize=(12,8))
vbp.plot.barh()
plt.xlabel('Volume')
plt.ylabel('Price')
plt.title(f'Volume by Price Chart for {NAME} from {START} to {END}')
plt.show()
```

```python
### price by volume indiactor
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define a function for the price by volume indicator
def pbv(ticker, start, end, period):
    import yfinance as yf
    # Get the historical data for the ticker
    data = yf.download(ticker, start, end)
    # Calculate the volume-weighted average price (VWAP)
    vwap = (data["Close"] * data["Volume"]).rolling(period).sum() / data["Volume"].rolling(period).sum()
    # Calculate the price by volume (PBV)
    data["PBV"] = (data["Close"] / vwap) - 1
    # Return the PBV series
    return data["Close"], data["PBV"]

# Apply the PBV function with a 20-day period
PERIOD = 20
CLOSE, PBV = pbv(NAME, START, END, PERIOD)

A = PBV[-len(CLOSE) + PERIOD:]
print(A)
DO = np.quantile(A, .025)
UP = np.quantile(A, .975)
# Plot the PBV indicator along with the closing price
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(CLOSE, color="blue", label="Close")
plt.legend(loc="upper left")
plt.semilogy()
plt.subplot(2, 1, 2)
plt.plot(PBV, color="black", label=f"PBV: {PBV[-1]:.2f}")
plt.axhline(y=DO, linestyle="--", color="green", label=f"Q 0.025: {DO:.2f}")
plt.axhline(y=0.0, linestyle='--', color='cyan', label="Zero")
plt.axhline(y=UP, linestyle="--", color="red",label=f"Q 0.975: {UP:.2f}")
plt.legend(loc="upper left")
plt.title(f'Price by Volume Chart for {NAME} based on {PERIOD} Periods from {START} to {END}')
plt.show()
```

```python
# Define the ichimoku cloud indicator function
def ichimoku_cloud(symbol, start, end):
    # Import libraries
    import matplotlib.pyplot as plt
    import yfinance as yf
    # Download the historical data from yfinance
    data = yf.download(symbol, start, end)
    
    # Calculate the five lines of the indicator
    data['tenkan_sen'] = (data['High'].rolling(9).max() + data['Low'].rolling(9).min()) / 2 # Conversion line
    data['kijun_sen'] = (data['High'].rolling(26).max() + data['Low'].rolling(26).min()) / 2 # Base line
    data['senkou_span_a'] = ((data['tenkan_sen'] + data['kijun_sen']) / 2).shift(26) # Leading span A
    data['senkou_span_b'] = ((data['High'].rolling(52).max() + data['Low'].rolling(52).min()) / 2).shift(26) # Leading span B
    data['chikou_span'] = data['Close'].shift(-26) # Lagging span
    
    # Plot the indicator on a chart
    plt.figure(figsize=(12,8))
    plt.plot(data['Close'], label='Close', color='black') # Plot the closing price
    plt.plot(data['tenkan_sen'], label='Tenkan-sen', color='blue') # Plot the conversion line
    plt.plot(data['kijun_sen'], label='Kijun-sen', color='red') # Plot the base line
    plt.plot(data['chikou_span'], label='Chikou span', color='green') # Plot the lagging span
    plt.fill_between(data.index, data['senkou_span_a'], data['senkou_span_b'], alpha=0.3, label='Kumo', color='grey') # Plot the cloud area
    plt.legend(loc='upper left') # Add a legend
    plt.title(f'Ichimoku Cloud for {symbol} Chart from {start} to {end}') # Add a title
    plt.xticks(rotation=90); plt.xlabel("Date"); plt.ylabel("Price"); plt.semilogy();
    plt.show() # Show the plot
    return data['tenkan_sen'], data['kijun_sen'], data['senkou_span_a'], data['senkou_span_b']

# Example usage: plot the ichimoku cloud for AAPL from 2020-01-01 to 2021-07-01
tenkan_sen, kijun_sen, senkou_span_a, senkou_span_b = ichimoku_cloud(NAME, START, END)
```

```python
# Define a function to plot a  Up Tick and Down Tick Chart
def plot_tick(ticker, start, end):
    import matplotlib.pyplot as plt
    import yfinance as yf
    # Get the data from yfinance
    data = yf.download(ticker, start, end)
    # Create a figure and an axis
    plt.figure(figsize=(12,8))
    ax1 = plt.gca(); ax2 = ax1.twinx();
    # Set the title and the labels
    ax1.set_title(f"{ticker} Up Tick and Down Tick Chart from {start} to {end}")
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Price")
    ax1.plot(np.arange(len(data["Close"])), data["Close"], color='black')
    # Loop through the data and plot the ticks
    for i in range(0, len(data)):
        # Get the close prices for each day
        Open = data["Open"][i]
        close = data["Close"][i]
        low = data["Low"][i]
        high = data["High"][i]
        prev_close = data["Close"][i-1]
        # Get the date as a string
        date = data.index[i].strftime("%Y-%m-%d")
        # Determine the color and the marker of the tick
        if close > prev_close and prev_close > low and close > low and close > Open and Open > low:
            color = "green"
            marker = "^"
        elif close < prev_close and prev_close < high and close < high and close < Open and Open < high:
            color = "red"
            marker = "v"
        else:
            color = "black"
            marker = ""
        # Plot the tick
        ax2.scatter(date, close, color=color, marker=marker)     
    # Show the plot
    plt.xticks(rotation=90); ax1.semilogy();ax2.semilogy();
    plt.grid()
    plt.show()

    
plot_tick(NAME, START, END)      
```

```python
########################################################################### 

def HISROGRAM_BOXPLOT(X, LOG_NORMAL, WEIGHT, PROBABILITY_RITO, BINS, HISTO_COLOR,LABEL):
    import numpy as np
    import matplotlib.pyplot as plt
    X = np.array(X)
    if LOG_NORMAL == 1:
        X = np.exp(X)
    print("-------------------------")
    from scipy.stats import skew, kurtosis
    MINIMUM = np.min(X)
    MAXIMUM = np.max(X)
    #MODE = max(set(X), key=list(X).count)
    MEDIAN = np.quantile(X, .50)#q2
    MEAN = np.mean(X)
    STD = np.std(X)
    q1 = np.quantile(X, .25)
    q3 = np.quantile(X, .75)
    SKEW = skew(X)
    KURT = kurtosis(X)
    #SKEW = (MEAN - MODE) / STD
    #KURT = (np.mean((X - MEAN)**4) / STD**4)
    # Estimate confidence intervals of the output variable
    lower_bound = np.quantile(X, .05)
    upper_bound = np.quantile(X, .95)
    print("Box-Chart Datas: ")
    print(f'Minimum: {MINIMUM:.4e}')
    print(f'First quartile: {q1:.4e}')
    #print(f'Mode: {MODE:.4f}')
    print(f'Median: {MEDIAN:.4e}')
    print(f'Mean: {MEAN:.4e}')
    print(f'Std: {STD:.4e}')
    print(f'Third quartile: {q3:.4e}')
    print(f'Maximum: {MAXIMUM :.4e}')
    print(f'Skewness: {skew(X) :.4e}')
    print(f'kurtosis: {kurtosis(X) :.4e}')
    print(f"90% Confidence Interval: ({lower_bound:.4e}, {upper_bound:.4e})")
    print("-------------------------")

    plt.figure(figsize=(10,6))
    if WEIGHT == 1:
        # calculate the bin edges and counts for the histogram
        counts, y = np.histogram(X, bins=BINS)
        # calculate the weights of each bar as percent of total
        weight = abs(counts) / abs(counts).sum() * 100

        # plot the histogram with weights
        plt.figure(figsize=(10,6))
        # use y as bins argument for plt.hist()
        plt.hist(y[1:], bins=BINS, weights=weight, color=HISTO_COLOR)
        plt.xlabel(LABEL)
        plt.ylabel('Percent (%)')
        plt.title('Weighted Histogram')
        plt.show()
        
    if WEIGHT == 0:
        # Plot histogram of data
        count, bins, ignored = plt.hist(X, bins=BINS, color=HISTO_COLOR, density=True, align='mid')#, edgecolor="black"

        # Plot lognormal PDF
        x = np.linspace(min(bins), max(bins), 10000)
        pdf = (np.exp(-(x - MEAN)**2 / (2 * STD**2)) / (STD * np.sqrt(2 * np.pi)))
        plt.plot(x, pdf, linewidth=2, color='r', label="Normal PDF")

        # Plot vertical lines for risk measures
        plt.axvline(q1, color="black", linestyle="--", label=f"Quantile 0.25: {q1:.4e}")
        plt.axvline(MEDIAN, color="green", linestyle="--", label=f"Median: {MEDIAN:.4e}")
        plt.axvline(q3, color="black", linestyle="--", label=f"Quantile 0.75: {q3:.4e}")
        #plt.axvline(MODE, color="purple", linestyle="--", label=f"Mode: {MODE:.4e}")
        plt.axvline(MEAN, color="red", linestyle="--", label=f"Mean: {MEAN:.4e}")
        plt.axvline(MEAN-STD, color="blue", linestyle="--", label=f"Mean-Std: {MEAN-STD:.4e}")
        plt.axvline(MEAN+STD, color="blue", linestyle="--", label=f"Mean+Std: {MEAN+STD:.4e}")
        plt.xlabel(LABEL)
        plt.ylabel("Frequency")
        prob = np.sum(X > 0) / len(X)
        plt.title(f"Histogram - Probability of Positive {LABEL} is {100*prob:.2f} %")
        plt.legend()
        #plt.grid()
        plt.show()

    #Plot boxplot with outliers
    plt.figure(figsize=(10,6))
    plt.boxplot(X, vert=0)
    # Write the quartile data on the chart
    #plt.text(q1, 1.05, f" Q1: {q1:.4f}")
    #plt.text(MEDIAN, 1.1, f" Q2: {MEDIAN:.4f}")
    #plt.text(q3, 1.05, f" Q3: {q3:.4f}")
    #plt.text(MODE, 1.15, f" Mode: {MODE:.4f}")
    
    #plt.text(MEAN, 0.9, f" Mean: {MEAN:.4f}")
    #plt.text(MEAN-STD, 0.9, f" Mean-Std: {MEAN-STD:.4f}")
    #plt.text(MEAN+STD, 0.9, f" Mean+Std: {MEAN+STD:.4f}")
    plt.scatter(q1, 1, color="cyan", s=200, label=f"Q1: {q1:.4e}")
    plt.scatter(MEDIAN, 1, color="lime", s=200, label=f"Median: {MEDIAN:.4e}")
    plt.scatter(q3, 1, color="purple", s=200, label=f"Q3:  {q3:.4e}")
    plt.scatter(MEAN, 1, color="red", marker="+", s=200, label=f"Mean: {MEAN:.4e}")
    plt.scatter(MEAN-STD, 1, color="green", marker="X", s=200, label=f"Mean-Std: {MEAN-STD:.4e}")
    plt.scatter(MEAN+STD, 1, color="blue", marker="*", s=200, label=f"Mean+Std:  {MEAN+STD:.4e}")
    plt.xlabel(LABEL)
    plt.ylabel("Data")
    
    # Quantile for Probability
    q = np.quantile(X, PROBABILITY_RITO)
    # Mean Quantile for Probability (Conditional Value) 
    mq = np.mean(X[X <= q])
    
    plt.title(f"Boxplot of {LABEL} - Mean {LABEL} Quantile for Probability of {100*PROBABILITY_RITO:.2f} % is {mq:.4f} ")
    plt.legend()
    plt.grid()
    plt.show()
    

HISROGRAM_BOXPLOT(prices, LOG_NORMAL=0, WEIGHT=0, PROBABILITY_RITO=0.05, BINS=100, HISTO_COLOR='orange', LABEL='Prices')
```

```python
def MEAN_CONFIDENCE_INTERVAL(data, confidence=0.95, LABEL='true'):
    import numpy as np
    import scipy.stats
    # convert data to numpy array
    a = np.array(data)
    # get the sample size
    n = len(a)
    # calculate the sample mean and standard deviation
    mean, sigma = np.mean(a), np.std(a)
    # calculate the margin of error using the normal distribution
    h = sigma * scipy.stats.norm.ppf((1 + confidence) / 2.0)
    # return the mean and the lower and upper bounds of the confidence interval
    LOW = mean-h
    HIGH = mean+h
    print(f'{LABEL}\t\t{LOW:.5f}\t\t{mean:.5f}\t\t{HIGH:.5f}')
    return mean, LOW, HIGH

print('\t\t\tCONFIDENCE INTERVAL FOR DIFFERENT PERIODS')
print('========================================================================')
print(f'\tCHANGE\t\tLOW\t\t\tMEAN\t\tHIGH')
MEAN_CONFIDENCE_INTERVAL(changes, confidence=0.5, LABEL='changes per day')
MEAN_CONFIDENCE_INTERVAL(changes_7, confidence=0.5, LABEL='changes 7 days')
MEAN_CONFIDENCE_INTERVAL(changes_30, confidence=0.5, LABEL='changes 30 days')
MEAN_CONFIDENCE_INTERVAL(changes_90, confidence=0.5, LABEL='changes 90 days')
MEAN_CONFIDENCE_INTERVAL(changes_220, confidence=0.5, LABEL='changes 220 days')

plt.figure(figsize=(10,6))
# Plot the histogram
plt.hist(changes, bins=50, alpha=0.5)
plt.hist(changes_7, bins=50, alpha=0.5)
plt.hist(changes_30, bins=50, alpha=0.5)
plt.hist(changes_90, bins=50, alpha=0.5)
plt.hist(changes_220, bins=50, alpha=0.5)
plt.xlabel("Changes per Day")
plt.ylabel("sample")
plt.title("Histogram of Sample for Changes per Day in Different Periods")
plt.grid()
plt.legend(['[1] Change 1 Time', '[2] Change 7 Time','[3] Change 30 Time','[4] Change 90 Time','[5] Change 220 Time'])
plt.show()


plt.figure(figsize=(10,6))
# Plot the boxplot with patch_artist=True and boxprops
plt.boxplot(changes, positions=[1], vert=0)
plt.boxplot(changes_7, positions=[2], vert=0) 
plt.boxplot(changes_30, positions=[3], vert=0)
plt.boxplot(changes_90, positions=[4], vert=0)
plt.boxplot(changes_220, positions=[5], vert=0)   
plt.xlabel("Changes per Day")
plt.ylabel("sample")
plt.title("Box Plot of Sample for Changes per Day in Different Periods")
plt.grid()
plt.legend(['[1] Change 1 Time', '[2] Change 7 Time','[3] Change 30 Time','[4] Change 90 Time','[5] Change 220 Time'])
plt.show()
```

```python
# Define a function to calculate the quantile bands
def quantile_bands(data, window, q):
    # Calculate the rolling quantiles
    lower = data.rolling(window).quantile(q[0])
    middle = data.rolling(window).quantile(q[1])
    upper = data.rolling(window).quantile(q[2])
    ratio = (data - lower) / (upper - lower)
    # Return a dataframe with the bands
    return lower, middle, upper, ratio

# Calculate the 20-day quantile bands with 0.10, 0.50, and 0.90 quantiles
PERIOD = 20 
lower, middle, upper, ratio = quantile_bands(prices, PERIOD, [0.10, 0.50, 0.90])

# Plot the price and the bands
plt.figure(figsize=(10, 6))
plt.plot(prices, label=f'Price: {prices[-1]:.2f}', color='black')
plt.plot(lower, label=f'Lower band: {lower[-1]:.2f}', color='lime')
plt.plot(middle, label=f'Middle band:  {middle[-1]:.2f}', color='red')
plt.plot(upper, label=f'Upper band:  {upper[-1]:.2f}', color='lime')
plt.title('Quantile Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.semilogy()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(ratio, color="black", label=f"Ratio: {ratio[-1]:.3f}")
plt.axhline(y=0, linestyle="--", color="green", label=f"Lower limit: {0:.3f}")
plt.axhline(y=0.5, linestyle='--', color='cyan', label=f"Mid limit: {0.5:.3f}")
plt.axhline(y=1, linestyle="--", color="red",label=f"Upper limit: {1.00:.3f}")
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.title('Quantile Bands Ratio')
plt.legend(loc="upper left")
plt.show()
```

```python
###########################################################################
################ Calculate Linear Regression Channel Indictor

def LRC_INDICATOR(prices, period):
    import numpy as np
    import matplotlib.pyplot as plt
    # calculate the x and y values for the linear regression
    x = np.arange(period)
    y = prices[:period]
    # calculate the slope and intercept of the best fit line
    slope, intercept = np.polyfit(x, y, 1)
    # calculate the LRC values for the window
    lrc = slope * x + intercept
    # initialize the LRC list with the first window values
    lrc_list = list(lrc)
    # loop through the remaining prices and update the LRC values
    for i in range(period, len(prices)):
        # shift the x and y values by one
        x = x + 1
        y = prices[i-period+1:i+1]
        # recalculate the slope and intercept
        slope, intercept = np.polyfit(x, y, 1)
        # append the latest LRC value to the list
        lrc_list.append(slope * x[-1] + intercept)
    # return the LRC list
    return lrc_list

# call the function with the generated prices and a period
LRC_20 = LRC_INDICATOR(prices, 20)
LRC_50 = LRC_INDICATOR(prices, 50)
LRC_100 = LRC_INDICATOR(prices, 100)
LRC_200 = LRC_INDICATOR(prices, 200)
LRC_400 = LRC_INDICATOR(prices, 400)

# plot the prices and the LRC indicator
plt.figure(figsize=(10, 6))
plt.plot(LL, prices, label="Prices", color="black")
plt.plot(LL, LRC_20, label=f"LRC 20 : {LRC_20[-1]:.2f}")
plt.plot(LL, LRC_50, label=f'LRC 50 : {LRC_50[-1]:.2f}')
plt.plot(LL, LRC_100, label=f'LRC 100 : {LRC_100[-1]:.2f}')
plt.plot(LL, LRC_200, label=f'LRC 200 : {LRC_200[-1]:.2f}')
plt.plot(LL, LRC_400, label=f'LRC 400 : {LRC_400[-1]:.2f}')
plt.title("Linear Regression Channel Indictor")
plt.legend()
plt.semilogy()
plt.show()
```

```python
###########################################################################
################ Calculate Zig Zag Indictor line

def Zig_Zag(prices, threshold):
    # Initialize the zig zag points list with the first price
    zz = [prices[0]]
    # Initialize the trend direction
    trend = None
    # Loop through the prices from index 1 to the end
    for i in range(1, len(prices)):
        # Calculate the percentage change from the last zig zag point
        change = (prices[i] - zz[-1]) / zz[-1]
        # If the trend direction is not set yet
        if trend is None:
            # If the change is greater than the threshold, set the trend to up and append the price as a zig zag point
            if change > threshold:
                trend = "up"
                zz.append(prices[i])
            # If the change is less than the negative threshold, set the trend to down and append the price as a zig zag point
            elif change < -threshold:
                trend = "down"
                zz.append(prices[i])
        # If the trend direction is up
        elif trend == "up":
            # If the change is less than the negative threshold, change the trend to down and append the price as a zig zag point
            if change < -threshold:
                trend = "down"
                zz.append(prices[i])
            # If the change is greater than zero and the price is higher than the last zig zag point, update the last zig zag point to the current price
            elif change > 0 and prices[i] > zz[-1]:
                zz[-1] = prices[i]
        # If the trend direction is down
        elif trend == "down":
            # If the change is greater than the threshold, change the trend to up and append the price as a zig zag point
            if change > threshold:
                trend = "up"
                zz.append(prices[i])
            # If the change is less than zero and the price is lower than the last zig zag point, update the last zig zag point to the current price
            elif change < 0 and prices[i] < zz[-1]:
                zz[-1] = prices[i]
    return zz


# Calculate the zig zag points with a 5% threshold
zz = Zig_Zag(prices, 0.05)

# Create a list of numbers from 0 to the length of prices
time = list(range(len(prices)))
zz_time = [i for i in range(len(prices)) if prices[i] in zz]

# Plot the prices and the Zig Zag
plt.figure(figsize=(10, 6)) 
plt.plot(LL, prices, label="Prices", color="black")
plt.plot(zz_time, zz, label="Zig Zag", color="red")#, marker="o"
plt.title("Price Zig Zag Line")
plt.xlabel("Time")
plt.ylabel("Price")
plt.semilogy()
plt.legend()
plt.show()

```

```python
###########################################################################
### FIND MAXIMUM AND MINIMUM PIVOTS ON CHARTS
def Find_Pivots(prices, window):
    # Find the local maxima and minima of the prices within a given window
    peaks = []
    troughs = []
    for i in range(window, len(prices) - window):
        if prices[i] == max(prices[i-window:i+window+1]):
            peaks.append((i, prices[i]))
        elif prices[i] == min(prices[i-window:i+window+1]):
            troughs.append((i, prices[i]))
    
    # Connect the peaks and troughs with lines and find the intersections
    intersections = []
    for i in range(len(peaks) - 1):
        for j in range(len(troughs) - 1):
            # Calculate the slopes and intercepts of the lines
            m1 = (peaks[i+1][1] - peaks[i][1]) / (peaks[i+1][0] - peaks[i][0])
            b1 = peaks[i][1] - m1 * peaks[i][0]
            m2 = (troughs[j+1][1] - troughs[j][1]) / (troughs[j+1][0] - troughs[j][0])
            b2 = troughs[j][1] - m2 * troughs[j][0]
            
            # Check if the lines intersect within the window
            if m1 != m2:
                x = (b2 - b1) / (m1 - m2)
                y = m1 * x + b1
                if x > peaks[i][0] and x < peaks[i+1][0] and x > troughs[j][0] and x < troughs[j+1][0]:
                    intersections.append((x, y))
    
    # Plot the prices, peaks, troughs and intersections
    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(len(prices)), prices, color='black')
    #plt.scatter([p[0] for p in peaks], [p[1] for p in peaks], color='red', label='Max. Pivots')
    #plt.scatter([t[0] for t in troughs], [t[1] for t in troughs], color='green', label='Min. Pivots')
    #plt.scatter([i[0] for i in intersections], [i[1] for i in intersections], color='blue')
    plt.plot([p[0] for p in peaks], [p[1] for p in peaks], color='red', marker='o', label='Max. Pivots')
    plt.plot([t[0] for t in troughs], [t[1] for t in troughs], color='green', marker='o', label='Min. Pivots')
    #plt.plot([i[0] for i in intersections], [i[1] for i in intersections], marker='o', color='blue')
    plt.title("Minimum and Maximum Pivots")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.semilogy()
    plt.legend()
    plt.show()

Find_Pivots(prices, 20)
```

```python
###########################################################################

def Plot_Trend_Line(prices):
    import matplotlib.pyplot as plt
    # Find the peak and trough indices of the prices
    peaks = [i for i in range(1, len(prices)-1) if prices[i] > prices[i-1] and prices[i] > prices[i+1]]
    troughs = [i for i in range(1, len(prices)-1) if prices[i] < prices[i-1] and prices[i] < prices[i+1]]
    # Find the slope and intercept of the trend lines
    slope = (prices[peaks[-1]] - prices[peaks[0]]) / (peaks[-1] - peaks[0])
    intercept = prices[peaks[0]] - slope * peaks[0]
    # Plot the prices and the trend lines
    plt.figure(figsize=(10,6))
    plt.plot(np.arange(len(prices)), prices, label="Prices", color="black")
    plt.plot([slope * i + intercept for i in range(len(prices))], label="Trend line", color="red")
    plt.legend()
    plt.title("Trading line")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.show()
    
Plot_Trend_Line(prices)
```

```python
###########################################################################
##############  rsi indictor 

def RSI_INDICATOR(prices, periods, ma_period, k):
    import numpy as np
    # Calculate the price changes
    changes = np.diff(prices)
    
    # Separate the positive and negative changes
    gains = np.where(changes > 0, changes, 0)
    losses = np.where(changes < 0, -changes, 0)
    
    # Calculate the exponential moving averages of gains and losses
    avg_gain = np.empty(len(prices))
    avg_loss = np.empty(len(prices))
    
    avg_gain[:periods] = np.mean(gains[:periods-1])
    avg_loss[:periods] = np.mean(losses[:periods-1])
    
    for i in range(periods, len(prices)):
        avg_gain[i] = (avg_gain[i-1]*(periods-1) + gains[i-1])/periods
        avg_loss[i] = (avg_loss[i-1]*(periods-1) + losses[i-1])/periods
    
    
    # Calculate the RSI
    rsi = 100 * (avg_gain / (avg_gain + avg_loss))
    rsi_ma = np.empty(len(prices))
    rsi_std = np.empty(len(prices))
    lower_band = np.empty(len(prices))
    upper_band = np.empty(len(prices))
    ratio = np.empty(len(prices))
    rsi_ma[:periods] = np.mean(rsi[:periods-1])
    rsi_std[:periods] = np.std(rsi[:periods-1])
    for i in range(periods, len(prices)):# moving average of rsi
        rsi_ma[i] = ( rsi_ma[i-1]*(periods-1) + rsi[i-1])/periods
        rsi_std[i] = np.std(rsi[i-periods+1:i+1])
        
    lower_band = rsi_ma - k * rsi_std
    upper_band = rsi_ma + k * rsi_std
    
    for i in range(periods + ma_period, len(prices)):
        ratio[i] = (rsi[i] - lower_band[i]) / (upper_band[i] - lower_band[i])
     
    return rsi, rsi_ma, lower_band, upper_band, ratio

##############  mfi indictor 

def MFI_INDICATOR(prices, volume, periods, ma_period, k):
    import numpy as np
    # Calculate the raw money flows
    cf = np.diff(prices) * volume[1:]
    # Calculate the positive and negative money flows
    pcf = np.where(cf > 0, cf, 0)
    ncf = np.where(cf < 0, -cf, 0)
    
    # Calculate the money flow ratios
    money_flow_ratio = np.empty(len(prices))
    positive_sum = np.empty(len(prices))
    negative_sum = np.empty(len(prices))
    for i in range(periods-1, len(prices)-1):
        positive_sum = np.sum(pcf[i-periods+1:i+1])
        negative_sum = np.sum(ncf[i-periods+1:i+1])
        if negative_sum == 0:
            money_flow_ratio[i+1] = np.inf
        else:
            money_flow_ratio[i+1] = positive_sum / negative_sum

    # Calculate the MFI
    mfi = 100 * (money_flow_ratio / (1 + money_flow_ratio))

    # Calculate the moving average and standard deviation of MFI
    mfi_ma = np.empty(len(prices))
    mfi_std = np.empty(len(prices))

    mfi_ma[:periods] = np.mean(mfi[:periods-1])
    mfi_std[:periods] = np.std(mfi[:periods-1])

    for i in range(periods, len(prices)):
        mfi_ma[i] = (mfi_ma[i-1]*(periods-1) + mfi[i-1])/periods
        mfi_std[i] = np.std(mfi[i-periods+1:i+1])
        
    # Calculate the lower and upper bands
    lower_band = mfi_ma - k * mfi_std
    upper_band = mfi_ma + k * mfi_std

    # Calculate the ratio
    ratio = np.empty(len(prices))

    for i in range(periods + ma_period, len(prices)):
        ratio[i] = (mfi[i] - lower_band[i]) / (upper_band[i] - lower_band[i])
     
    return mfi, mfi_ma, lower_band, upper_band, ratio

##############  cci indictor 

def CCI_INDICATOR(prices, period):
    import numpy as np
    prices = np.array(prices).reshape(-1, 1)
    # Calculate the typical price
    tp = np.mean(prices, axis=1)
    # Calculate the moving average of the typical price
    ma = np.convolve(tp, np.ones(period)/period, mode='valid')
    # Calculate the mean deviation
    md = np.zeros_like(ma)
    for i in range(period, len(tp)):
        md[i-period] = np.mean(np.abs(tp[i-period:i] - ma[i-period]))
    # Calculate the CCI
    cci = (tp[period-1:] - ma) / (0.015 * md)
    return cci


##############  bollinger bands indictor 

def Bollinger_Bands(prices, periods, k):
    import numpy as np
    # Calculate the simple moving average
    sma = np.convolve(prices, np.ones(periods)/periods, mode='valid')
    
    # Calculate the standard deviation
    std = np.std(prices[:periods-1])
    
    # Initialize the upper and lower bands
    ma_band = np.empty(len(prices))
    upper_band = np.empty(len(prices))
    lower_band = np.empty(len(prices))
    
    ma_band[:periods-1] = sma[0]
    upper_band[:periods-1] = sma[0] + k*std
    lower_band[:periods-1] = sma[0] - k*std
    ratio = np.zeros(len(prices))
    # Update the bands for each period
    for i in range(periods-1, len(prices)):
        std = np.std(prices[i-periods+1:i+1])
        ma_band[i] = sma[i-periods+1]
        upper_band[i] = sma[i-periods+1] + k*std
        lower_band[i] = sma[i-periods+1] - k*std
        ratio[i] = (prices[i] - lower_band[i]) / (upper_band[i] - lower_band[i])
    
    return ma_band, upper_band, lower_band, ratio

############## quantile bands indictor

# Define a function to calculate the quantile bands
def quantile_bands(data, window, q):
    # Calculate the rolling quantiles
    lower = data.rolling(window).quantile(q[0])
    middle = data.rolling(window).quantile(q[1])
    upper = data.rolling(window).quantile(q[2])
    ratio = (data - lower) / (upper - lower)
    # Return a dataframe with the bands
    return lower, middle, upper, ratio

############## MACD indictor

def MACD(prices, short_period, long_period, signal_period):
    import numpy as np
    import pandas as pd # added this line
    short_ema = pd.Series(prices).ewm(span=short_period, adjust=False).mean() # changed this line
    long_ema = pd.Series(prices).ewm(span=long_period, adjust=False).mean() # changed this line
    macd_line = short_ema - long_ema
    signal_line = macd_line.ewm(span=signal_period, adjust=False).mean()
    macd_line = np.array(macd_line)# change array pandas to numpy
    signal_line = np.array(signal_line)# change array pandas to numpy
    return macd_line, signal_line

############## Momentum indictor
"""
def Momentum(changes, period):
    momentum = []
    for i in range(len(changes)):
        if i < period:
            momentum.append(0)
        else:
            momentum.append(sum(changes[i-period+1:i+1]))
    momentum = np.array(momentum)        
    return momentum
"""
def Momentum(prices, period):
    momentum = []
    for i in range(len(prices)):
        if i < period:
            momentum.append(prices[1] - prices[0])
        else:
            momentum.append(prices[i] - prices[i-period])
    momentum = np.array(momentum)        
    return momentum


##############  roc indictor 

def ROC_INDICATOR(prices, period):
    import numpy as np
    roc = np.zeros(len(prices))
    for i in range(period, len(prices)):
        roc[i] = (prices[i] - prices[i-period]) / prices[i-period]
    return roc

##############  Directional Movement Indicator

# Define DMI function
def DMI(symbol, start, end, n=14, n_ma=20, k=2):
    # Download historical data from yfinance
    data = yf.download(symbol, start=start, end=end)
    # Calculate the true range (TR)
    data['TR'] = data['High'] - data['Low']
    data['TR'] = data['TR'].where(data['TR'] > abs(data['Close'].shift(1) - data['High']), abs(data['Close'].shift(1) - data['High']))
    data['TR'] = data['TR'].where(data['TR'] > abs(data['Close'].shift(1) - data['Low']), abs(data['Close'].shift(1) - data['Low']))
    # Calculate the directional movement (+DM and -DM)
    data['+DM'] = data['High'] - data['High'].shift(1)
    data['-DM'] = data['Low'].shift(1) - data['Low']
    data['+DM'] = data['+DM'].where(data['+DM'] > 0, 0)
    data['-DM'] = data['-DM'].where(data['-DM'] > 0, 0)
    data['+DM'] = data['+DM'].where(data['+DM'] > data['-DM'], 0)
    data['-DM'] = data['-DM'].where(data['-DM'] > data['+DM'], 0)
    # Calculate the smoothed averages of TR, +DM, and -DM
    data['ATR'] = data['TR'].rolling(n).sum()
    data['+DM_smooth'] = data['+DM'].rolling(n).sum()
    data['-DM_smooth'] = data['-DM'].rolling(n).sum()
    # Calculate the +DI and -DI
    data['+DI'] = 100 * data['+DM_smooth'] / data['ATR']
    data['-DI'] = 100 * data['-DM_smooth'] / data['ATR']
    # Calculate the ADX
    data['DX'] = 100 * abs(data['+DI'] - data['-DI']) / (data['+DI'] + data['-DI'])
    data['ADX'] = data['DX'].rolling(n).mean()
    
    # Calculate the moving average and standard deviation of the ADX
    data['ADX_MA'] = data['DX'].rolling(n_ma).mean()
    data['ADX_STD'] = data['DX'].rolling(n_ma).std()
    MIN = data['ADX_MA'] - k * data['ADX_STD']
    MAX = data['ADX_MA'] + k * data['ADX_STD']
    data['DX_RATIO'] = (data['DX'] - MIN) / (MAX - MIN)
    # Return the DMI components
    return data[['+DI', '-DI', 'ADX', 'DX_RATIO']].dropna()

##############  z-score Indicator
# Calculate z-score function for period
def Z_Score_Indicator(x, period):
    import numpy as np
    z = np.zeros(len(x))
    # Calculate the rolling mean and standard deviation of x for period
    mean = x.rolling(period).mean()
    std = x.rolling(period).std()

    # Calculate the z-score for each element of x
    z = (x - mean) / std
    z = z[period:]

    # Return the z-score
    return z


# Calculate the  RSI and MFI and bollinger bands and Roc and MACD and Momentum
rsi, rsi_ma, rsi_lower_band, rsi_upper_band, rsi_ratio = RSI_INDICATOR(prices, 14, 20, 2)
rsi_low = np.quantile(rsi, 0.025)
rsi_mid = np.quantile(rsi, 0.50)
rsi_high = np.quantile(rsi, 0.975)

mfi, mfi_ma, mfi_lower_band, mfi_upper_band, mfi_ratio = MFI_INDICATOR(prices, volume, 20, 20, 2)
mfi_low = np.quantile(mfi, 0.025)
mfi_mid = np.quantile(mfi, 0.50)
mfi_high = np.quantile(mfi, 0.975)

cci = CCI_INDICATOR(prices, 20)
cci_low = np.quantile(cci, 0.025)
cci_mid = np.quantile(cci, 0.50)
cci_high = np.quantile(cci, 0.975)

ma_band, upper_band, lower_band, bb_ratio = Bollinger_Bands(prices, 20, 2) 
ma_band, upper_band, lower_band, bb_ratio02 = Bollinger_Bands(bb_ratio, 20, 2) 

lower, middle, upper, qq_ratio = quantile_bands(prices, 20, [0.10, 0.50, 0.90])

roc = 100*ROC_INDICATOR(prices, 20)
roc_low = np.quantile(roc, 0.025)
roc_mid = np.quantile(roc, 0.50)
roc_high = np.quantile(roc, 0.975)
roc_ratio = (roc[-1] - roc_low) / (roc_high - roc_low)

macd_line, signal_line = MACD(prices, 12, 26, 9)
macd = macd_line; 
macd_low = np.quantile(macd, 0.025)
macd_mid = np.quantile(macd, .50)
macd_high = np.quantile(macd, 0.975)
macd_ratio = (macd_line[-1] - macd_low) / (macd_high - macd_low)


mom = Momentum(prices, 20)
mom_low = np.quantile(mom, 0.025)
mom_mid = np.quantile(mom, 0.50)
mom_high = np.quantile(mom, 0.975)
mom_ratio = (mom[-1] - mom_low) / (mom_high - mom_low)


dmi = DMI(NAME, START, END, n=20, n_ma=20, k=2)


z_score = Z_Score_Indicator(prices, period = 20)
z_score_low = np.quantile(z_score, 0.025)
z_score_mid = np.quantile(z_score, 0.50)
z_score_high = np.quantile(z_score, 0.975)


# Plot the datas
plt.figure(figsize=(18,24))
plt.subplot(11,1,1)
plt.plot(100*bb_ratio, label=f'BB%: {100*bb_ratio[-1]:.2f}', color='black')
plt.axhline(0, color='green', linestyle='--')
plt.axhline(y=50.0, linestyle='--', color='cyan')
plt.axhline(100, color='red', linestyle='--')
plt.legend(loc="lower left")
plt.title('Bollinger Bands')

plt.subplot(11,1,2)
plt.plot(100*bb_ratio02, label=f'BB%: {100*bb_ratio02[-1]:.2f}', color='black')
plt.axhline(0, color='green', linestyle='--')
plt.axhline(y=50.0, linestyle='--', color='cyan')
plt.axhline(100, color='red', linestyle='--')
plt.legend(loc="lower left")
plt.title('Bollinger Bands of Bollinger Bands')

plt.subplot(11,1,3)
plt.plot(100*qq_ratio, label=f'QB%: {100*qq_ratio[-1]:.2f}', color='black')
plt.axhline(0, color='green', linestyle='--')
plt.axhline(y=50.0, linestyle='--', color='cyan')
plt.axhline(100, color='red', linestyle='--')
plt.legend(loc="lower left")
plt.title('Quntile Bands')

plt.subplot(11,1,4)
plt.plot(np.arange(len(dmi['DX_RATIO'])), 100*dmi['DX_RATIO'], label=f"DX%: {100*dmi['DX_RATIO'][-1]:.2f}", color='black')
plt.axhline(y=0.0, linestyle="--", color="green")
plt.axhline(y=50.0, linestyle='--', color='cyan')
plt.axhline(y=100.0, linestyle="--", color="red")
plt.title(f'Directional Movement Indicator for {NAME}')
plt.legend(loc="lower left")

"""
plt.plot(prices, label='Price', color='black')
plt.plot(ma_band, label='ma', color='red')
plt.plot(upper_band, label='Upper band', color='orange')
plt.plot(lower_band, label='Lower band', color='orange')
"""

plt.subplot(11,1,5)
plt.plot(rsi, label=f'RSI: {rsi[-1]:.2f}', color='black')
plt.plot(rsi_ma, label=f'RSI MA: {rsi_ma[-1]:.2f}', color='red')
plt.plot(rsi_lower_band, label=f'RSI MA - K*STD: {rsi_lower_band[-1]:.2f}', color='lime')
plt.plot(rsi_upper_band, label=f'RSI MA + K*STD: {rsi_upper_band[-1]:.2f}', color='lime')
plt.axhline(rsi_low, color='green', linestyle='--')
plt.axhline(rsi_mid, color='cyan', linestyle='--')
plt.axhline(rsi_high, color='red', linestyle='--')
plt.text(0, rsi_low, f"LOW: {rsi_low:.2f}") # write the equation of the line on the plot
plt.text(0, rsi_mid, f"MID: {rsi_mid:.2f}") # write the equation of the line on the plot
plt.text(0, rsi_high, f"HIGH: {rsi_high:.2f}") # write the equation of the line on the plot
plt.title('Relative Strength Index')
plt.legend(loc="lower left")

plt.subplot(11,1,6)
plt.plot(mfi, label=f'MFI: {mfi[-1]:.2f}', color='black')
plt.plot(mfi_ma, label=f'MFI MA: {mfi_ma[-1]:.2f}', color='red')
plt.plot(mfi_lower_band, label=f'MFI MA - K*STD: {mfi_lower_band[-1]:.2f}', color='lime')
plt.plot(mfi_upper_band, label=f'MFI MA + K*STD: {mfi_upper_band[-1]:.2f}', color='lime')
plt.axhline(mfi_low, color='green', linestyle='--')
plt.axhline(mfi_mid, color='cyan', linestyle='--')
plt.axhline(mfi_high, color='red', linestyle='--')
plt.text(0, mfi_low, f"LOW: {mfi_low:.2f}") # write the equation of the line on the plot
plt.text(0, mfi_mid, f"MID: {mfi_mid:.2f}") # write the equation of the line on the plot
plt.text(0, mfi_high, f"HIGH: {mfi_high:.2f}") # write the equation of the line on the plot
plt.title('Money Flow Index')
plt.legend(loc="lower left")


plt.subplot(11,1,7)
plt.plot(roc, color="black", label=f"ROC: {roc[-1]:.2f}")
plt.axhline(roc_low, color='green', linestyle='--')
plt.axhline(roc_mid, color='cyan', linestyle='--')
plt.axhline(roc_high, color='red', linestyle='--')
plt.text(0, roc_low, f"LOW: {roc_low:.2f}") # write the equation of the line on the plot
plt.text(0, roc_mid, f"MID: {roc_mid:.2f}") # write the equation of the line on the plot
plt.text(0, roc_high, f"HIGH: {roc_high:.2f}") # write the equation of the line on the plot
plt.title("Rate of Change")
plt.legend(loc="lower left")

plt.subplot(11,1,8)
plt.plot(cci, color="black", label=f"CCI: {cci[-1]:.2f}")
plt.axhline(cci_low, color='green', linestyle='--')
plt.axhline(cci_mid, color='cyan', linestyle='--')
plt.axhline(cci_high, color='red', linestyle='--')
plt.text(0, cci_low, f"LOW: {cci_low:.2f}") # write the equation of the line on the plot
plt.text(0, cci_mid, f"MID: {cci_mid:.2f}") # write the equation of the line on the plot
plt.text(0, cci_high, f"HIGH: {cci_high:.2f}") # write the equation of the line on the plot
plt.title("Commodity Channel Index")
plt.legend(loc="lower left")

plt.subplot(11,1,9)
plt.plot(macd_line, color="black")
plt.plot(signal_line, color="red")
plt.axhline(macd_low, color='green', linestyle='--')
plt.axhline(macd_mid, color='cyan', linestyle='--')
plt.axhline(macd_high, color='red', linestyle='--')
plt.text(0, macd_low, f"LOW: {macd_low:.2f}") # write the equation of the line on the plot
plt.text(0, macd_mid, f"MID: {macd_mid:.2f}") # write the equation of the line on the plot
plt.text(0, macd_high, f"HIGH: {macd_high:.2f}") # write the equation of the line on the plot
plt.title(f"MACD")
plt.legend(loc="lower left")

plt.subplot(11,1,10)
plt.plot(mom, color="black", label=f"Momentum: {mom[-1]:.2f}")
plt.axhline(mom_low, color='green', linestyle='--')
plt.axhline(mom_mid, color='cyan', linestyle='--')
plt.axhline(mom_high, color='red', linestyle='--')
plt.text(0, mom_low, f"LOW: {mom_low:.2f}") # write the equation of the line on the plot
plt.text(0, mom_mid, f"MID: {mom_mid:.2f}") # write the equation of the line on the plot
plt.text(0, mom_high, f"HIGH: {mom_high:.2f}") # write the equation of the line on the plot
plt.title("Momentum")
plt.legend(loc="lower left")


plt.subplot(11,1,11)
plt.plot(np.arange(len(z_score)), z_score, color="black", label=f"z score: {z_score[-1]:.2f}")
plt.axhline(z_score_low, color='green', linestyle='--')
plt.axhline(z_score_mid, color='cyan', linestyle='--')
plt.axhline(z_score_high, color='red', linestyle='--')
plt.text(0, z_score_low, f"LOW: {z_score_low:.2f}") # write the equation of the line on the plot
plt.text(0, z_score_mid, f"MID: {z_score_mid:.2f}") # write the equation of the line on the plot
plt.text(0, z_score_high, f"HIGH: {z_score_high:.2f}") # write the equation of the line on the plot
plt.title("z score")
plt.legend(loc="lower left")

plt.show()
```

```python
###########################################################################
##############       CALCULATE RETURN AND STD FOR EACH PERIOD

def RETURN_STD(changes, period):
    num_days = len(changes)
    import numpy as np
    # Calculate the return and standard deviation of prices for each period days using numpy functions
    returns = [np.prod(1 + np.array(changes[i:i+period])) - 1 for i in range(0, num_days, period)] # List of returns for each period days
    stds = [np.std(np.array(changes[i:i+period])) for i in range(0, num_days, period)] # List of standard deviations for each period days
    returns = [x for x in returns if str(x) != 'nan' and x != 0]
    stds = [x for x in stds if str(x) != 'nan' and x != 0]
    returns = np.array(returns)
    stds = np.array(stds)
    return returns, stds

def LINEAR_REGRESSION_02(x, y, XL='xlabel', YL='ylabel', TI='title', SX=0, SY=0):
    # Import libraries
    import numpy as np
    # Convert x,y to a numpy array
    x = np.array(x)
    y = np.array(y)

    # Add a column of ones to X for the intercept term
    X = np.c_[np.ones(x.shape[0]), x]

    # Solve the equation a x = b for x
    coeffs = np.linalg.lstsq(X, y, rcond=None)[0]

    # Print the coefficients
    print(f"Coefficients: {coeffs[1:]}") # coefficients
    print(f"Intercept: {coeffs[0]}") # intercept
    # print(f'A * X + B : {coeffs[1:]:.2e} X + {coeffs[0]:.2e}')

    plt.figure(figsize=(10,6))
    plt.scatter(x, y, label='Datas')
    
    if SX == 1:
        plt.semilogx()
    if SY == 1:
        plt.semilogy()    
        
    #plt.semilogy()
    plt.xlabel(XL)
    plt.ylabel(YL)
    plt.title(TI)
    # write the equation of the line on the plot
    r_squared = (np.corrcoef(x, y)[0][1]) ** 2
    plt.plot(X[:, 1], X @ coeffs,color='red', label=f"y = {coeffs[1]:.4f}x + {coeffs[0]:.4f} -  R^2 = {r_squared:.2f}") # plot the fitted line
    #plt.text(np.mean(x), np.mean(y), f"y = {coeffs[1]:.2f}x + {coeffs[0]:.2f} -  R^2 = {r_squared:.2f}") 
    plt.legend()
    plt.show()    
    
PERIOD = 10 #period must greater than 2 
returns, stds = RETURN_STD(changes, PERIOD)
LINEAR_REGRESSION_02(stds, returns, XL="Standard Deviations", YL="Return", TI=f"Standard Deviations and Returnsbased on period {PERIOD}", SX=0, SY=0)

def HISTOGRAM02(X1, X2, XL1, XL2, BINS):
    # Plot histogram of X1 distribution
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.hist(X1, bins=BINS, edgecolor="black")
    plt.xlabel(XL1)
    plt.ylabel("Frequency")
    plt.title(f"{XL1} distribution")
    #plt.semilogy()
    # Plot histogram of X2 distribution 
    plt.subplot(1, 2, 2)
    plt.hist(X2, bins=BINS, edgecolor="black")
    plt.xlabel(XL2)
    plt.ylabel("Frequency")
    plt.title(f"{XL2} distribution")
    #plt.semilogy()
    plt.tight_layout()
    plt.show()

def BOXPLOT02(X, Y, XL, YL):
    import matplotlib.pyplot as plt
    import numpy as np
    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    # Plot the boxplot for returns on the first subplot
    ax1.boxplot(Y)
    ax1.set_title(YL)
    ax1.set_ylabel(YL)
    # Calculate the quartiles for returns
    q1_y = np.quantile(Y, 0.25)
    q2_y = np.quantile(Y, 0.5)
    q3_y = np.quantile(Y, 0.75)
    # Write the quartile data on the chart
    ax1.text(1.1, q1_y, f" Q1: {q1_y:.4f}")
    ax1.text(1.1, q2_y, f" Q2: {q2_y:.4f}")
    ax1.text(1.1, q3_y, f" Q3: {q3_y:.4f}")
    # Plot the boxplot for stds on the second subplot
    ax2.boxplot(X)
    ax2.set_title(XL)
    ax2.set_ylabel(XL)
    # Calculate the quartiles for stds
    q1_x = np.quantile(X, 0.25)
    q2_x = np.quantile(X, 0.5)
    q3_x = np.quantile(X, 0.75)
    # Write the quartile data on the chart
    ax2.text(1.1, q1_x, f" Q1: {q1_x:.4f}")
    ax2.text(1.1, q2_x, f" Q2: {q2_x:.4f}")
    ax2.text(1.1, q3_x, f" Q3: {q3_x:.4f}")
    # Show the figure
    plt.show()
    print(f"MEDIAN {YL} TO {XL} IS {100*(q2_y/q2_x):.4f} (%)")
    
HISTOGRAM02(returns,stds, BINS=50, XL1="Return", XL2="Standard Deviations") 
BOXPLOT02(stds, returns, XL="Standard Deviations", YL="Return")    
```

```python
###########################################################################
##############       CALCULATE BETA AND CORRELATION COEFFICIENT


def ALPHA(prices, prices_index):
    # write alpha coefficient function between (prices, prices_index)
    import numpy as np
    # calculate the returns of prices and prices_index
    returns = [prices[i+1] / prices[i] - 1 for i in range(len(prices) - 1)]
    returns_index = [prices_index[i+1] / prices_index[i] - 1 for i in range(len(prices_index) - 1)]
    # calculate the beta of prices with respect to prices_index
    beta = np.cov(returns_index, returns)[0][1] / np.var(returns_index)
    # calculate the expected return of prices given beta and market return
    expected_return = np.mean(returns_index) * beta  
    # calculate the alpha coefficient as the difference between actual and expected return
    return np.mean(returns) - expected_return


def BETA(A, B):
    import numpy as np
    # calculate beta function (slope of A and B)
    return np.cov(A, B)[0][1] / np.var(B)

beta = BETA(prices_index, prices)

def CORRELATION(A, B):
    import numpy as np
    # calculate correlation function (correlation of A and B)
    return np.corrcoef(A, B)[0][1]

# calculate correlation function (correlation of prices_index and prices)
alpha = ALPHA(changes, changes_index)
beta = BETA(changes_index, changes)
correlation = CORRELATION(changes_index, changes)

# print the alpha coefficient
print("--------------------------------------"), 
print(f"        Alpha coefficient: {alpha:.4f}")
print(f"         Beta coefficient: {beta:.4f}")
print(f"  Correlation coefficient: {CORRELATION(changes_index, changes):.4f}")
print("--------------------------------------")

# plot prices and prices_index in 2 y axis separately
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()
ax1.plot(prices_index, color="blue", label="Index")
ax2.plot(prices, color="black", label="Prices")
ax1.set_xlabel("Time")
ax1.set_ylabel("Index")
ax2.set_ylabel("Price")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
ax1.semilogy(); ax2.semilogy();
plt.title(f"Prices and Index Chart - Alpha: {alpha:.4f} , Beta: {beta:.4f} , Correlation: {correlation:.4f}")
plt.show()
```

```python
###########################################################################

def Market_Trend(prices):
    import numpy as np
    import scipy.stats as sps
    prices = np.array(prices) #this is a numpy array
    #calculate returns
    returns = np.log(prices[1:]/prices[:-1])
    returns = returns[returns != 0]
    n = len(returns)
    signs = np.sign(returns)
    runs = signs[1:] - signs[:-1]
    observed_runs = np.count_nonzero(runs == 2) + np.count_nonzero(runs == -2) + 1
    positive_returns = np.count_nonzero(signs == 1)
    negative_returns = np.count_nonzero(signs == -1)
    expected_runs = 2*positive_returns*negative_returns/n + 1
    stdev_runs = (expected_runs*(expected_runs - 1)/(n - 1))**(1/2)
    z_stat = (observed_runs - expected_runs)/stdev_runs
    p_value = 2*(1 - sps.norm.cdf(abs(z_stat)))
    if p_value > 0.1:
        print('market is efficient')
    elif z_stat > 0:
        print('market is mean-reverting')
    else:
        print('market is trending')
    print(f'observed_runs: {observed_runs} \nexpected runs: {int(expected_runs)} \nstdev runs: {round(stdev_runs, 2)} \nz stat: {round(z_stat, 4)} \np-value: {round(p_value, 4)}')
    return observed_runs, int(expected_runs), round(stdev_runs, 2), round(z_stat, 4), round(p_value, 4)

def Distribution_Fitting(prices):
    import numpy as np
    import scipy.stats as sps
    prices = np.array(prices) #this is a numpy array
    #calculate returns
    returns = np.log(prices[1:]/prices[:-1])
    #sort returns
    returns = np.sort(returns)
    #construct empirical distribution function
    edf = np.arange(1, len(returns)+1)/len(returns)
    #construct normal distribution function
    mean = np.average(returns)
    std = np.std(returns)
    normal_cdf = sps.norm.cdf(returns, mean, std)
    #plot the distribution functions
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(returns, edf,label='edf', color='red')
    plt.plot(returns, normal_cdf,label='cdf', color='black')
    plt.legend()
    #apply the Kolmogorov-Smirnov test
    p_value = np.exp(-max(abs(edf - normal_cdf))**2*len(returns))
    if p_value < 0.1: 
        print('normal distribution fits the data: p-value '+str(round(p_value, 4)))
    else:
        print('normal distribution does not fit the data: p-value '+str(round(p_value, 4)))
        
    #construct cauchy distribution function
    median = np.quantile(returns, 0.5)
    iq_range = (np.quantile(returns, 0.75) - np.quantile(returns, 0.25))/2
    cauchy_cdf = 1/np.pi*np.arctan((returns - median)/iq_range) + 1/2
    #plot the distribution functions
    plt.subplot(1, 2, 2)
    plt.plot(returns, cauchy_cdf,label='cauchy cdf', color='red')
    plt.plot(returns, normal_cdf,label='cdf', color='black')
    plt.tight_layout()
    plt.legend()
    plt.show()
    #apply the Kolmogorov-Smirnov test
    p_value = np.exp(-max(abs(edf - cauchy_cdf))**2*len(returns))
    if p_value < 0.1:
         print('Cauchy distribution fits the data: p-value '+str(round(p_value, 4)))
    else:
         print('Cauchy distribution does not fit the data: p-value '+str(round(p_value, 4)))
        
    return 0

Market_Trend(prices)
Distribution_Fitting(prices)
```

```python
###########################################################################
#### VOLATILITY BASED ON EACH PERIODS

def Calculate_Volatility(prices, changes, periods):
    # prices: a list of prices generated by Generate_Prices
    # changes: a list of percentage changes generated by Generate_Prices
    # periods: a list of integers representing the number of days for each period
    # returns: a list of volatilities for each period

    volatilities = [] # initialize an empty list to store the volatilities
    start_index = 0 # initialize the starting index of each period
    for period in periods: # loop through each period
        end_index = start_index + period# calculate the ending index of each period
        if end_index > len(prices): # check if the ending index exceeds the length of prices
            break # stop the loop if so
        period_prices = prices[start_index:end_index] # slice the prices for the current period
        period_changes = changes[start_index:end_index] # slice the changes for the current period
        mean_change = sum(period_changes) / len(period_changes) # calculate the mean percentage change for the current period
        #period_changes = 1 + np.array(changes[start_index:end_index]) # slice the changes for the current period
        #mean_change = (np.prod(period_changes) ** (1/len(period_changes))) - 1 # calculate the mean percentage change for the current period
        squared_deviations = [(change - mean_change)**2 for change in period_changes] # calculate the squared deviations from the mean change for each day
        variance = sum(squared_deviations) / len(squared_deviations) # calculate the variance of percentage changes for the current period
        volatility = np.sqrt(variance) * np.sqrt(252) # calculate the annualized volatility for the current period
        volatilities.append(volatility) # append the volatility to the list
        start_index = end_index # update the starting index for the next period
    return volatilities # return the list of volatilities


periods = np.array([2, 5, 10, 20, 50, 100, 200, 400])
periods_r = ["2", "5", "10", "20", "50", "100", "200", "400"]


# calculate volatilities
VOL = Calculate_Volatility(prices, changes, periods)
q1 = np.quantile(VOL, 0.25)
q2 = np.quantile(VOL, 0.5)
q3 = np.quantile(VOL, 0.75)    
#plot the original Period and the Volatility
plt.figure(figsize=(10, 6))
plt.bar(periods_r, VOL, color="yellow", edgecolor="black")
plt.axhline(q1, color='green', linestyle='--')
plt.axhline(q2, color='cyan', linestyle='--')
plt.axhline(q3, color='red', linestyle='--')
plt.text(0, q1, f" Quantile 0.25: {q1:.2f}") # write the equation of the line on the plot
plt.text(0, q2, f" Quantile 0.50: {q2:.2f}") # write the equation of the line on the plot
plt.text(0, q3, f" Quantile 0.75: {q3:.2f}") # write the equation of the line on the plot
plt.xlabel("Time")
plt.ylabel("Annualized Volatility")
plt.title("Annualized Volatility of Price change in Different Periods")
#plt.semilogy()
plt.show()    

# print the results
for i in range(len(periods)):
    print(f"Period {periods[i]} Volatility {100*VOL[i]:.2f} %")
    
print('\n')    
```

```python
"""
write model to prediction next 7 days price with using MARKOV CHAIN and plot the result
"""
# Import libraries
import random
import numpy as np
import matplotlib.pyplot as plt

# Define the function to generate prices
def Generate_Prices(start_price, Volatility, Direction_Bias, num_days):
    import random
    prices = [start_price]
    changes = [0]
    for i in range(num_days-1):
        Var = random.uniform(-1, 1)*random.uniform(0, 1)*Volatility
        price = prices[-1] *(1 + Var + Direction_Bias)
        prices.append(price)
        changes.append(prices[-1] / prices[i-1] - 1)
    return prices, changes

# Define the function to predict the next 7 days price using Markov chain
def Predict_Prices(prices, changes, num_days):
    # Calculate the transition matrix
    states = [-1, 0, 1] # -1: down, 0: no change, 1: up
    matrix = np.zeros((3, 3)) # 3x3 matrix to store the probabilities
    for i in range(len(changes) - 1):
        # Find the current and next state based on the sign of the change
        current_state = np.sign(changes[i])
        next_state = np.sign(changes[i+1])
        # Update the matrix by adding 1 to the corresponding cell
        matrix[states.index(current_state), states.index(next_state)] += 1
    # Normalize the matrix by dividing each row by its sum
    matrix = matrix / matrix.sum(axis=1, keepdims=True)
    
    # Predict the next 7 days price using Markov chain
    predictions = [prices[-1]] # Initialize the predictions list with the last price
    for i in range(num_days):
        # Find the current state based on the last price change
        current_state = np.sign(changes[-1])
        # Sample the next state from the transition matrix
        next_state = np.random.choice(states, p=matrix[states.index(current_state)])
        # Calculate the next price based on the average change of the next state
        avg_change = np.mean([c for c in changes if np.sign(c) == next_state])
        next_price = predictions[-1] * (1 + avg_change)
        # Append the next price to the predictions list
        predictions.append(next_price)
        # Update the changes list with the new change
        changes.append(next_price / predictions[-2] - 1)
    return predictions

# Generate the prices and changes
start_price = 100
Volatility = 0.05
Direction_Bias = 0.0001 
num_days = 1000
NEXT_PREDICTED_DAYS = 30
prices, changes = Generate_Prices(start_price, Volatility, Direction_Bias, num_days)    

# Predict the next 7 days price
predictions = Predict_Prices(prices, changes, NEXT_PREDICTED_DAYS)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(prices, label='Historical prices', color='black')
plt.plot(range(num_days, num_days + NEXT_PREDICTED_DAYS + 1), predictions, label='Predicted prices', color='red')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Price prediction using Markov chain')
plt.legend()
plt.semilogy()
plt.show()
```

```python
###### Stock Risk Analysis using Weighted RSI and Correlation Coefficients
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def ROC(data):# calculate rate of change
    roc = [0]
    for i in range(1, len(data)):
        roc.append((data[i] - data[i-1]) / data[i-1]) 
    return 100*roc

def GET_DATA_YFINANCE(NAME, start=START, end=END , interval= "1d"):
    stocks = yf.download(NAME, start=START, end=END , interval= "1d")#EURUSD=X
    prices = stocks['Close']
    changes = ROC(prices)
    #print(stocks)
    return prices, changes


def moving_average_correlation(x, y, window=20):
    x = pd.Series(x)
    y = pd.Series(y)
    corr = x.rolling(window).corr(y)
    return corr  

def RSI_INDICATOR(prices, periods, ma_period, k):
    import numpy as np
    # Calculate the price changes
    changes = np.diff(prices)
    
    # Separate the positive and negative changes
    gains = np.where(changes > 0, changes, 0)
    losses = np.where(changes < 0, -changes, 0)
    
    # Calculate the exponential moving averages of gains and losses
    avg_gain = np.empty(len(prices))
    avg_loss = np.empty(len(prices))
    
    avg_gain[:periods] = np.mean(gains[:periods-1])
    avg_loss[:periods] = np.mean(losses[:periods-1])
    
    for i in range(periods, len(prices)):
        avg_gain[i] = (avg_gain[i-1]*(periods-1) + gains[i-1])/periods
        avg_loss[i] = (avg_loss[i-1]*(periods-1) + losses[i-1])/periods
    
    
    # Calculate the RSI
    rsi = 100 * (avg_gain / (avg_gain + avg_loss))
    rsi_ma = np.empty(len(prices))
    rsi_std = np.empty(len(prices))
    lower_band = np.empty(len(prices))
    upper_band = np.empty(len(prices))
    ratio = np.empty(len(prices))
    rsi_ma[:periods] = np.mean(rsi[:periods-1])
    rsi_std[:periods] = np.std(rsi[:periods-1])
    for i in range(periods, len(prices)):# moving average of rsi
        rsi_ma[i] = ( rsi_ma[i-1]*(periods-1) + rsi[i-1])/periods
        rsi_std[i] = np.std(rsi[i-periods+1:i+1])
        
    lower_band = rsi_ma - k * rsi_std
    upper_band = rsi_ma + k * rsi_std
    
    for i in range(periods + ma_period, len(prices)):
        ratio[i] = (rsi[i] - lower_band[i]) / (upper_band[i] - lower_band[i])
     
    return rsi, rsi_ma, lower_band, upper_band, ratio

######## YAHOO FINANCE DATA
START = "2023-01-01"
END = "2023-06-01"
NAME_STOCK = 'AAPL'#'CHFUSD=X''FB','AMZN','NFLX','GOOG', 'AAPL','NFLX'
NAME_INDEX01 = '^GSPC'#'GBPUSD=X'
NAME_INDEX02 = '^NDX'#'GBPUSD=X'
NAME_INDEX03 = 'DOW'#'GBPUSD=X'


### LOAD STOCK DATA
prices_STOCK, changes_STOCK = GET_DATA_YFINANCE(NAME_STOCK, start=START, end=END , interval= "1d")    
prices_INDEX01, changes_INDEX01 = GET_DATA_YFINANCE(NAME_INDEX01, start=START, end=END , interval= "1d")  
prices_INDEX02, changes_INDEX02 = GET_DATA_YFINANCE(NAME_INDEX02, start=START, end=END , interval= "1d") 
prices_INDEX03, changes_INDEX03 = GET_DATA_YFINANCE(NAME_INDEX03, start=START, end=END , interval= "1d")  

### CALCULATE CORRELATION COEFFICIENT 
CORRE01 = moving_average_correlation(prices_STOCK, prices_INDEX01)
CORRE02 = moving_average_correlation(prices_STOCK, prices_INDEX02)
CORRE03 = moving_average_correlation(prices_STOCK, prices_INDEX03)
# Corrected Correlation
CORRE01 = 0.5 * (CORRE01 + 1)
CORRE02 = 0.5 * (CORRE02 + 1)
CORRE03 = 0.5 * (CORRE03 + 1)
COE01 = CORRE01[-1] / (CORRE01[-1] + CORRE02[-1] + CORRE03[-1])
COE02 = CORRE02[-1] / (CORRE01[-1] + CORRE02[-1] + CORRE03[-1])
COE03 = CORRE03[-1] / (CORRE01[-1] + CORRE02[-1] + CORRE03[-1])
print("WEIGHT OF CORRELATION COEFFICIENT: ",COE01 , COE02, COE03)

### LOAD RSI INDIACTOR
PERIOD = 50
MA_PERIOD = 20
K = 2
_, _, _, _, rsi_ratio01 = RSI_INDICATOR(prices_INDEX01, PERIOD, MA_PERIOD, K)
_, _, _, _, rsi_ratio02 = RSI_INDICATOR(prices_INDEX02, PERIOD, MA_PERIOD, K)
_, _, _, _, rsi_ratio03 = RSI_INDICATOR(prices_INDEX03, PERIOD, MA_PERIOD, K)
_, _, _, _, rsi_ratio04 = RSI_INDICATOR(prices_STOCK, PERIOD, MA_PERIOD, K)

WEIGHTED_RSI = COE01 * rsi_ratio01[-1] + COE02 * rsi_ratio02[-1] + COE03 * rsi_ratio03[-1]
print("RSI: ",rsi_ratio01[-1], rsi_ratio02[-1], rsi_ratio03[-1], rsi_ratio04[-1])
print("WEIGHTED RSI: ",WEIGHTED_RSI)
if (rsi_ratio04[-1] > WEIGHTED_RSI or rsi_ratio04[-1] >= 1):
    print("High Risk Postion, It is recommended waiting!")
    
# Plot the results
plt.figure(figsize=(10,6))
plt.plot(rsi_ratio01[-MA_PERIOD:], label=f'RSI {NAME_INDEX01}: {rsi_ratio01[-1]:.2f}')
plt.plot(rsi_ratio02[-MA_PERIOD:], label=f'RSI {NAME_INDEX02}: {rsi_ratio02[-1]:.2f}')
plt.plot(rsi_ratio03[-MA_PERIOD:], label=f'RSI {NAME_INDEX03}: {rsi_ratio03[-1]:.2f}')
plt.plot(rsi_ratio04[-MA_PERIOD:], label=f'RSI {NAME_STOCK}: {rsi_ratio04[-1]:.2f}')
plt.axhline(y=WEIGHTED_RSI, color='r', linestyle='--', label=f'Weighted Last RSI: {WEIGHTED_RSI:.2f}')
plt.axhline(y=0.0, color='cyan', linestyle='--', label=f'Minimum RSI: {0.0:.2f}')
plt.axhline(y=1.0, color='lime', linestyle='--', label=f'Maximum RSI: {1.0:.2f}')
plt.title('Stock Risk Analysis using Weighted RSI Channel (0 until 1) and Correlation Coefficients')
plt.legend()
plt.grid()
#plt.semilogy()
plt.show()    
```

```python
# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define DMI function
def DMI(symbol, start, end, n=14, k=2):
    # Download historical data from yfinance
    data = yf.download(symbol, start=start, end=end)
    # Calculate the true range (TR)
    data['TR'] = data['High'] - data['Low']
    data['TR'] = data['TR'].where(data['TR'] > abs(data['Close'].shift(1) - data['High']), abs(data['Close'].shift(1) - data['High']))
    data['TR'] = data['TR'].where(data['TR'] > abs(data['Close'].shift(1) - data['Low']), abs(data['Close'].shift(1) - data['Low']))
    # Calculate the directional movement (+DM and -DM)
    data['+DM'] = data['High'] - data['High'].shift(1)
    data['-DM'] = data['Low'].shift(1) - data['Low']
    data['+DM'] = data['+DM'].where(data['+DM'] > 0, 0)
    data['-DM'] = data['-DM'].where(data['-DM'] > 0, 0)
    data['+DM'] = data['+DM'].where(data['+DM'] > data['-DM'], 0)
    data['-DM'] = data['-DM'].where(data['-DM'] > data['+DM'], 0)
    # Calculate the smoothed averages of TR, +DM, and -DM
    data['ATR'] = data['TR'].rolling(n).sum()
    data['+DM_smooth'] = data['+DM'].rolling(n).sum()
    data['-DM_smooth'] = data['-DM'].rolling(n).sum()
    # Calculate the +DI and -DI
    data['+DI'] = 100 * data['+DM_smooth'] / data['ATR']
    data['-DI'] = 100 * data['-DM_smooth'] / data['ATR']
    # Calculate the ADX
    data['DX'] = 100 * abs(data['+DI'] - data['-DI']) / (data['+DI'] + data['-DI'])
    data['ADX'] = data['DX'].rolling(n).mean()
    
    # Calculate the moving average and standard deviation of the ADX
    data['ADX_MA'] = data['DX'].rolling(20).mean()
    data['ADX_STD'] = data['DX'].rolling(20).std()
    MIN = data['ADX_MA'] - k * data['ADX_STD']
    MAX = data['ADX_MA'] + k * data['ADX_STD']
    data['ADX_RATIO'] = (data['DX'] - MIN) / (MAX - MIN)
    # Return the DMI components
    return data[['+DI', '-DI', 'ADX', 'ADX_RATIO']].dropna()

# Example usage: plot DMI for Apple from Jan 2023 to Aug 2023
symbol = 'AAPL'
start = '2023-01-01'
end = '2023-08-05'
n = 14 # default period for DMI calculation
dmi = DMI(symbol, start, end, n)
plt.figure(figsize=(12,3))
plt.plot(dmi['ADX_RATIO'], label=f"ADX RATIO: {dmi['ADX_RATIO'][-1]:.2f}", color='black')
plt.axhline(y=0.0, linestyle="--", color="purple", label=f"Lower Limit: {0.00}")
plt.axhline(y=1.0, linestyle="--", color="purple", label=f"Upper Limit {1.00}")
plt.xlabel("Time")
plt.ylabel("ADX")
plt.title(f'Directional Movement Indicator for {symbol}')
plt.legend()
plt.grid()
plt.show()
```

```python
"""
Financial Risk Management with Markov Chain Monte Carlo (MCMC) methods:
It is a very important topic for investors, traders, and regulators. Financial risk management is the process of identifying,
measuring, and managing the potential losses from various sources of uncertainty in the financial markets. One of the tools
that can be used for financial risk management is Markov Chain Monte Carlo (MCMC) methods.
MCMC methods are a class of algorithms that generate samples from a probability distribution using a Markov chain.
A Markov chain is a stochastic process that has the property of memorylessness, meaning that the next state only depends on
the current state and not on the previous states. By running a Markov chain for a long time, we can obtain samples that
approximate the target distribution. MCMC methods can be used to estimate various quantities of interest, such as risk
measures, risk allocations, parameter values, and model evidence.
One of the applications of MCMC methods in financial risk management is to estimate systemic risk allocations. Systemic
risk is the risk of collapse or instability of the entire financial system due to the interconnections and interdependencies
among its components. Systemic risk allocations are ways of assigning the responsibility or contribution of each component to
the overall systemic risk. For example, we can use conditional Value-at-Risk (CoVaR), conditional expected shortfall (CoES),
VaR contributions, and range VaR (RVaR) contributions as systemic risk allocations. These allocations depend on the conditional
distribution of the losses of each component given a crisis event, which is a rare event that indicates a high level of systemic
risk. MCMC methods can help us sample from this conditional distribution and estimate the allocations efficiently.
To illustrate how MCMC methods can be used for financial risk management, I have written some Python code that uses yfinance
library to get data from Yahoo Finance, MCMC methods to estimate CoVaR and CoES allocations for some selected stocks, and
matplotlib library to plot the results. You can see the code below:
"""
# Import libraries
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Define stocks and crisis event
stocks = ["AAPL", "MSFT", "AMZN", "GOOG"] # Apple, Microsoft, Amazon, Google, Facebook, "FB"
crisis = 0.05 # Crisis event is when the total loss exceeds 5%

# Get daily returns data from Yahoo Finance
data = yf.download(stocks, start="2020-01-01", end="2023-05-01")["Adj Close"].pct_change().dropna()
returns = data.to_numpy()

# Define MCMC parameters
n_iter = 10000 # Number of iterations
n_burn = 1000 # Number of burn-in samples
n_thin = 10 # Thinning factor

# Define functions for MCMC sampling
def log_target(x):
    # Logarithm of target density (multivariate normal with mean zero and covariance matrix)
    n = len(x)
    sigma = np.cov(returns.T)
    return -0.5 * n * np.log(2 * np.pi) - 0.5 * np.log(np.linalg.det(sigma)) - 0.5 * x.T @ np.linalg.inv(sigma) @ x

def log_proposal(x, y):
    # Logarithm of proposal density (multivariate normal with mean x and variance 0.01)
    n = len(x)
    return -0.5 * n * np.log(2 * np.pi) - 0.5 * n * np.log(0.01) - 0.5 * ((y - x) / 0.1) ** 2

def metropolis_hastings(x):
    # Metropolis-Hastings algorithm
    y = x + 0.1 * np.random.randn(len(x)) # Generate proposal from normal distribution
    A = np.exp(log_target(y) + log_proposal(y, x) - log_target(x) - log_proposal(x, y))
    alpha = min(1, min(A)) # Acceptance probability
    u = np.random.rand() # Uniform random number
    if u < alpha: # Accept proposal with probability alpha
        return y
    else: # Reject proposal and keep current state
        return x

# Initialize Markov chain
x = np.zeros(len(stocks)) # Initial state (zero vector)

# Run MCMC sampling
samples = [] # List to store samples
for i in range(n_iter):
    x = metropolis_hastings(x) # Update state using Metropolis-Hastings algorithm
    if i >= n_burn and i % n_thin == 0: # Store sample after burn-in and thinning
        samples.append(x)

# Convert samples to numpy array
samples = np.array(samples)

# Define functions for risk measures and allocations
def var(x, alpha):
    # Value-at-Risk at level alpha
    return -np.quantile(x, alpha)

def es(x, alpha):
    # Expected shortfall at level alpha
    return -np.mean(x[x <= -var(x, alpha)])

def covar(x, y, alpha):
    # Conditional Value-at-Risk of x given y at level alpha
    return -np.mean(x[y <= -var(y, alpha)])

def coes(x, y, alpha):
    # Conditional expected shortfall of x given y at level alpha
    return -np.mean(x[x <= -covar(x, y, alpha)])

# Estimate risk measures and allocations from samples
var_total = var(np.sum(samples, axis=1), crisis) # VaR of total loss
es_total = es(np.sum(samples, axis=1), crisis) # ES of total loss
covar_alloc = np.array([covar(samples[:, i], np.sum(samples, axis=1), crisis) for i in range(len(stocks))]) # CoVaR allocations
coes_alloc = np.array([coes(samples[:, i], np.sum(samples, axis=1), crisis) for i in range(len(stocks))]) # CoES allocations

# Print results
print("VaR of total loss:", var_total)
print("ES of total loss:", es_total)
print("CoVaR allocations:", covar_alloc)
print("CoES allocations:", coes_alloc)

# Plot results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(stocks, covar_alloc)
plt.title("CoVaR allocations")
plt.xlabel("Stocks")
plt.ylabel("CoVaR")
plt.grid()
plt.subplot(1, 2, 2)
plt.bar(stocks, coes_alloc)
plt.title("CoES allocations")
plt.xlabel("Stocks")
plt.ylabel("CoES")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Import yfinance library
import yfinance as yf

# Download apple stocks data from 2020 to 2021
data = yf.download("AAPL", start="2020-01-01", end="2021-01-01")

# Calculate z-score function for period 20
def z_score_20(x, period):
    # Calculate the rolling mean and standard deviation of x for period 20
    mean = x.rolling(period).mean()
    std = x.rolling(period).std()

    # Calculate the z-score for each element of x
    z = (x - mean) / std

    # Return the z-score
    return z

# Apply z-score function to the closing price column of data
data["z_score_20"] = z_score_20(data["Close"], period = 100)

# Plot the result
import matplotlib.pyplot as plt
#plt.style.use('dark_background')
plt.figure(figsize=(12,4))
plt.plot(data["z_score_20"])
plt.title("Z-score of Apple stocks closing price for period 20 from 2020 to 2021")
plt.xlabel("Date")
plt.ylabel("Z-score")
plt.show()
```

```python
### write python code and plot the result about Do stocks follow random walks? Wald and Wolfowitz (1940)

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Import yfinance library
import yfinance as yf

# Download apple stocks data from 2020 to 2021
data = yf.download("AAPL", start="2020-01-01", end="2021-01-01")

print(data)

# Calculate the daily returns
data["Return"] = data["Adj Close"].pct_change()

# Drop the first row with missing value
data = data.dropna()

# Define a function to count the number of runs
def count_runs(series):
    # Initialize the count and the previous value
    count = 0
    prev = None
    # Loop through the series
    for x in series:
        # If the current value is not equal to the previous value, increment the count and update the previous value
        if x != prev:
            count += 1
            prev = x
    # Return the count
    return count

# Define a function to perform the runs test
def runs_test(series, alpha=0.05):
    # Convert the series to a binary sequence
    series = np.sign(series)
    # Count the number of runs
    n_runs = count_runs(series)
    # Calculate the mean and variance of the number of runs under the null hypothesis
    n_pos = (series == 1).sum()
    n_neg = (series == -1).sum()
    n = n_pos + n_neg
    mean = 2 * n_pos * n_neg / n + 1
    var = 2 * n_pos * n_neg * (2 * n_pos * n_neg - n) / (n ** 2 * (n - 1))
    # Calculate the z-score and the p-value
    z = (n_runs - mean) / np.sqrt(var)
    p = 2 * stats.norm.sf(abs(z)) # Two-tailed test
    # Print the results
    print(f"Number of runs: {n_runs}")
    print(f"Mean: {mean}")
    print(f"Variance: {var}")
    print(f"Z-score: {z}")
    print(f"P-value: {p}")
    # Check if the p-value is less than the significance level
    if p < alpha:
        print(f"The p-value is less than {alpha}, so we reject the null hypothesis.")
        print(f"There is evidence of non-randomness in the series.")
    else:
        print(f"The p-value is greater than {alpha}, so we fail to reject the null hypothesis.")
        print(f"There is no evidence of non-randomness in the series.")

# Perform the runs test on the daily returns
runs_test(data["Return"])

# Plot the daily returns
plt.figure(figsize=(10, 6))
plt.plot(data["Return"], color="blue")
plt.title("Daily Returns of S&P 500")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```

```python
"""
The Profit Taking Index (PTI) is a technical indicator used in Elliott Wave analysis to assess the likelihood of a price reaching the end of a corrective wave. It is calculated by dividing the area of the corrective wave (typically Wave B) by the area of the prior impulse wave (typically Wave A).

A higher PTI value suggests that the corrective wave is more likely to complete, while a lower PTI value indicates that there is a greater probability of the corrective wave extending further. A PTI value of 35 is often considered a critical threshold, and a PTI above 35 is considered strong evidence that the corrective wave is nearing completion.

Here's the formula for calculating PTI:

PTI = (Area of Wave B) / (Area of Wave A)
The areas of the waves can be calculated using various methods, such as the Donchian Channel or the Chaikin A/D Line. However, it is important to note that the PTI is not a perfect indicator, and it should be used in conjunction with other Elliott Wave analysis tools to make trading decisions.

Here are some additional things to keep in mind when using the PTI:

The PTI is most effective when used on larger time frames, such as daily or weekly charts.
The PTI is less effective on shorter time frames, such as intraday charts, due to the increased volatility and noise.
The PTI can be used to identify potential divergences, which can signal a change in trend.
Overall, the Profit Taking Index is a useful tool for Elliott Wave traders, but it should not be used in isolation. By combining the PTI with other Elliott Wave analysis techniques, traders can improve their chances of making profitable trading decisions.

write python code and model Profit Taking Index indicator and plot it
"""
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define a function to calculate the Donchian Channel
def donchian_channel(data, window):
    # Calculate the upper and lower bands
    upper = data.rolling(window).max()
    lower = data.rolling(window).min()
    # Calculate the channel width
    width = upper - lower
    # Return the upper, lower, and width as a DataFrame
    return pd.DataFrame({'upper': upper, 'lower': lower, 'width': width})

# Define a function to calculate the Profit Taking Index
def profit_taking_index(data, window):
    # Calculate the Donchian Channel for the data
    dc = donchian_channel(data, window)
    # Initialize an empty list to store the PTI values
    pti = []
    # Loop through the data
    for i in range(len(data)):
        # If the current index is less than the window size, append None
        if i < window:
            pti.append(None)
        # Otherwise, calculate the PTI
        else:
            # Get the current price
            price = data[i]
            # Get the previous window prices
            prev_prices = data[i-window:i]
            # Get the previous window Donchian Channel values
            prev_dc = dc[i-window:i]
            # Calculate the area of Wave A
            area_a = np.trapz(prev_prices - prev_dc['lower'])
            # Calculate the area of Wave B
            area_b = np.trapz(price - prev_dc['upper'])
            # Calculate the PTI
            pti_value = (area_b / area_a) * 100
            # Append the PTI value to the list
            pti.append(pti_value)
    # Return the PTI values as a Series
    return pd.Series(pti, index=data.index, name='pti')

# Import yfinance library
import yfinance as yf

# Download apple stocks data from 2020 to 2021
data = yf.download('AAPL', start="2021-01-01", end="2023-09-01")# "EURUSD=X"'AAPL'

print(data)


# Calculate the PTI
pti = profit_taking_index(data['Adj Close'], window = 100)

# Plot the data and the PTI
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 6))
ax1.plot(data['Adj Close'], label='Price')
ax1.set_title('Sample Data')
ax1.legend()
ax1.semilogy()
ax2.plot(pti, color='black', label=f"Last PTI : {pti[-1]:.2f}")
ax2.set_title('Profit Taking Index')
ax2.axhline(0, color='cyan', linestyle='--', label=f'Threshold: 0')
ax2.axhline(35, color='red', linestyle='--', label=f'Threshold: 35')
#ax2.axhline(18, color='blue', linestyle='--', label='Threshold')
#ax2.axhline(0, color='cyan', linestyle='--', label='Threshold')
ax2.legend()
plt.show()
```

```python
# Author: Mostafa Shabani
# Date: 2024-09-01
# Source : Financial Academy

#Reminder: Make sure to install the required libraries before running the code:
#pip install pandas numpy statsmodels yfinance finpy_tse

import pandas as pd
import numpy as np
import statsmodels.api as sm
import yfinance as yf
import finpy_tse as fpy

def get_monthly_prices(stock_name, start_date, end_date):
    """
    Fetches monthly adjusted close prices for a given stock within a specified date range.
    """
    stock_data = fpy.Get_Price_History(stock=stock_name, start_date=start_date, end_date=end_date,
                                        show_weekday=True, adjust_price=True, double_date=True)
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    stock_data = stock_data.set_index('Date')
    stock_data = pd.DataFrame(stock_data['Adj Close'])
    stock_data = stock_data.resample('M').last()
    stock_data.index = stock_data.index.to_period('M')
    return stock_data

def get_Daily_prices(stock_name, start_date, end_date):
    """
    Fetches monthly adjusted close prices for a given stock within a specified date range.
    """
    stock_data = fpy.Get_Price_History(stock=stock_name, start_date=start_date, end_date=end_date,
                                        show_weekday=True, adjust_price=True, double_date=True)
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    stock_data = stock_data.set_index('Date')
    stock_data = pd.DataFrame(stock_data['Adj Close'])
    stock_data = stock_data.resample('D').last()
    stock_data.index = stock_data.index.to_period('D')
    return stock_data

def get_tedpix(start_date, end_date):
    """
    Fetches monthly TEDPIX data within a specified date range.
    """
    tedpix_data = fpy.Get_CWI_History(start_date=start_date, end_date=end_date,
                                       just_adj_close=False, double_date=True)
    tedpix_data['Date'] = pd.to_datetime(tedpix_data['Date'])
    tedpix_data = tedpix_data.set_index('Date')
    tedpix_data = pd.DataFrame(tedpix_data['Adj Close'])
    tedpix_data = tedpix_data.resample('M').last()
    tedpix_data.index = tedpix_data.index.to_period('M')
    return tedpix_data

def calculate_returns(data):
    """
    Calculates monthly returns.
    """
    returns = data.pct_change().dropna()
    return returns

def perform_regression(dependent_variable, independent_variable):
    """
    Performs regression analysis.
    """
    independent_variable = sm.add_constant(independent_variable)
    regression_model = sm.OLS(dependent_variable, independent_variable).fit()
    return regression_model

# Stock1:
Stock1 = get_monthly_prices(stock_name="", start_date="1396-01-01", end_date="1403-06-11")

# Stock2:
Stock2 = get_monthly_prices(stock_name="", start_date="1396-01-01", end_date="1403-06-11")

# TEDPIX (Tehran Exchange Dividend and Price Index)
TEDPIX = get_tedpix(start_date='1396-01-01', end_date='1403-06-11')

# Concatenating data
TSE = pd.concat([TEDPIX, Stock1, Stock2], axis=1)
TSE.columns = ['TSE', 'MADN', 'MSMI']

# Calculating returns
TSE_returns = calculate_returns(TSE)


colors = ['#CD5C5C', '#f4b5d0', '#0095b7']
TSE_returns.plot(figsize=(10, 6), color=colors)
plt.title('Monthly Returns of TEDPIX, Mavadan, and Fomeli')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.grid(True)
plt.show()
```
